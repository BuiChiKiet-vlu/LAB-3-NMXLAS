```
import numpy as np
import matplotlib.pylab as plt
import imageio.v2 as iio 

image_path = "XLB3/exercise/colorful-ripe-tropical-fruits.jpg"
bdata = iio.imread(image_path)

#Chọn trái đu đủ
bmg = bdata[276:757, 54:704].copy() #Chọn hình trái đu đủ từ ảnh gốc và tạo 1 bản copy để không bị ảnh hưởng từ hình gốc

#Tạo ảnh xanh lá 
ppyG = bmg[:, :, 1] + bmg[:, :, 2]
ppy_rgb = np.zeros_like(bmg) #Tạo một mảng ảnh màu toàn màu đen
ppy_rgb[:, :, 1] = np.clip(ppyG, 0, 255) #Dòng này đưa biến ppyG vào màu xanh lá, dùng np.clip để đảm bảo giá trị không vượt quá 255

#Chọn trái dưa hấu
bmg2 = bdata[272:1117, 1561:2108].copy() #Tương tự như phần chọn hình trái đu đủ

#Tạo ảnh xanh dương
#Về cách làm cũng tương tự với hình trái đu đủ nhưng đổi màu thành màu xanh dương 
wtmB = bmg2[:, :, 0] + bmg2[:, :, 1]
wtm_rgb = np.zeros_like(bmg2)
wtm_rgb[:, :, 2] = np.clip(wtmB, 0, 255)

print(bdata.shape)

#Hiển thị và lưu
iio.imsave("papaya.jpg", bmg)
plt.imshow(bmg)
plt.show()

iio.imsave("watermelon.jpg", bmg2)
plt.imshow(bmg2)
plt.show()

iio.imsave("papayaGreen.jpg", ppy_rgb)
plt.imshow(ppy_rgb)
plt.show()

iio.imsave("watermelonBlue.jpg", wtm_rgb)
plt.imshow(wtm_rgb)
plt.show()
```
